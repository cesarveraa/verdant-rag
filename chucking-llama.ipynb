```python
import PyPDF2

# Ruta al archivo PDF
pdf_path = 'libro1.pdf'

# Extraer texto del PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        for page in pdf_reader.pages:
            text += page.extract_text() if page.extract_text() else ""
    return text

agriculture_text = extract_text_from_pdf(pdf_path)
print(agriculture_text[:500])  # Muestra los primeros 500 caracteres
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Dividir el texto en fragmentos
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=0,
    length_function=len,
)

chunks = text_splitter.split_text(agriculture_text)

print(f"Se generaron {len(chunks)} fragmentos de texto.")
print(chunks[:5])  # Muestra los primeros 5 fragmentos para verificar
```

```python
from langchain.schema import Document

# Convertir los fragmentos (chunks) a objetos Document
documents = [Document(page_content=chunk) for chunk in chunks]

print(f"Se generaron {len(documents)} documentos.")
print(documents[:2])  # Muestra los primeros 2 documentos para verificar la conversión
```

```python
from sentence_transformers import SentenceTransformer
from langchain_community.vectorstores import FAISS
from langchain.embeddings.base import Embeddings  # Cambiar a la clase base genérica

# Cargar el modelo de embeddings locales con Sentence Transformers
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Crear una clase de Embeddings compatible con Langchain sin heredar de OpenAIEmbeddings
class LocalEmbeddings(Embeddings):
    def embed_documents(self, texts):
        """Convierte una lista de textos en una lista de vectores."""
        return embedding_model.encode(texts, convert_to_tensor=False).tolist()
    
    def embed_query(self, text):
        """Convierte una consulta en un vector."""
        return embedding_model.encode([text], convert_to_tensor=False).tolist()[0]

# Instanciar la clase de embeddings sin requerir API Key de OpenAI
embedding = LocalEmbeddings()

# Indexar los documentos utilizando FAISS con la clase de embeddings personalizada
semantic_chunk_vectorstore = FAISS.from_documents(documents, embedding=embedding)

# Configurar el recuperador semántico
semantic_chunk_retriever = semantic_chunk_vectorstore.as_retriever(search_kwargs={"k": 1})

print("Indexación con FAISS completada correctamente usando embeddings locales.")
```

```python
def ask_with_chunk_display(questions, retriever, llm_function):
    """
    Muestra el chunk exacto recuperado y la respuesta generada por el modelo para cada pregunta.
    
    Parámetros:
    - questions (list): Lista de preguntas a realizar.
    - retriever: El recuperador semántico configurado con FAISS.
    - llm_function (callable): La función que llama al modelo Llama 3.3 (SambaNova API).
    """
    for question in questions:
        print(f"\n🟢 Pregunta: {question}")
        
        # Recuperar el chunk semántico exacto desde FAISS
        retrieved_docs = retriever.get_relevant_documents(question)
        
        if retrieved_docs:
            # Mostrar el chunk exacto recuperado
            context = retrieved_docs[0].page_content
            print("\n🔍 Chunk Recuperado:")
            print(context[:500])  # Muestra los primeros 500 caracteres del chunk
            
            # Obtener la respuesta del modelo utilizando el chunk como contexto
            result = llm_function(question, context)
            print(f"\n🤖 Respuesta del Modelo:\n{result}")
        else:
            print("⚠️ No se recuperaron chunks relevantes.")

# Lista de preguntas a realizar
questions = [
    "What is crop rotation and why is it important?",
    "How can soil pH affect plant growth?",
    "What are effective water management practices in agriculture?"
]

# Llamar a la función adaptada
ask_with_chunk_display(questions, semantic_chunk_retriever, query_llama_sambanova)
```
